# Giving columns names to raw heart Dataset

```python
names = "age,sex,cp,trestbps,chol,fbs,restecg,thalach,exang,oldpeak,slope,ca,thal,RESULT"
```

```python
names = names.split(",")
```

```python
names
```

```python
len(names)
```

```python
data = pd.read_csv("heart_1.csv", names = names)
```

```python
data
```

```python
model = BayesianModel([("age", "sex"), ("sex", "cp"), ("cp", "RESULT")])
```

```python
model.fit(data)
```

```python
from pgmpy.inference import VariableElimination
infer = VariableElimination(model)
q = infer.query(variables=["RESULT"], evidence={"age":22})
```
