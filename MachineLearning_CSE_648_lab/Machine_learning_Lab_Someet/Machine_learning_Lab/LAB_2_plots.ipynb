```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
heart_d = pd.read_csv("heart-disease.csv")
```

```python
heart_d.info()
```

# Creating Histogram For heart_d['Age'] 

```python
fig,ax = plt.subplots(figsize=(10,10))
heart_d["age"].plot(kind="hist", title="age", bins=10);

```

# Creating Scatter plot for x and y variable to show relationship between two sets of data.

```python
fig, ax=plt.subplots(figsize=(10,5))
heart_d.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            ax=ax);
ax.set_xlim([45,100]) ##limits x axis
ax.set_ylim([200,700]); ##limits y axis
```

# Creating Pie Chart

```python
car_name = "toyota", "bmw", "nissan", "bugati"
car_p = [20,25,40,10]
```

# Creating bar chart

```python
# creating own Data and own key for bar chart
nut_butter_prices = {"almond butter": 10, 
                     "peanut butter":8, 
                     "cashew butter": 12 }  ##red is keys, numerical are values
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys() ,height = nut_butter_prices.values())
ax.set(title="nut butter prices", xlabel="Keys", ylabel="values");
```
